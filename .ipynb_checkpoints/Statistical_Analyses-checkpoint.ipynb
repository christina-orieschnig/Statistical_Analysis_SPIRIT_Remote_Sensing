```python
#####################################################################################################
### this script provides a statistical analysis of the evolution of different vegetation indices on irrigated 
### and not irrigated vineyards in the area of the canal du Gignac 
#####################################################################################################
```

```python
# import necessary packages 
from io import StringIO
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import datetime
import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
import math
from functools import reduce
import json 
import csv
import pandas as pd 
from pandas.io.json import json_normalize
import json
from datetime import datetime
#import pymannkendall as mk
#import pyhomogeneity as hg
from matplotlib.dates import date2num
from datetime import datetime
import matplotlib.ticker as ticker
import numpy as np
from scipy import stats
rng = np.random.default_rng()
```

```python
# import data 

# chrips 

chirps = pd.read_csv('ChirpsPrecipitation.csv')
chirps = chirps.drop(columns = ['system:index', '.geo']) 
chirps['date'] = pd.to_datetime(chirps['date'])
chirps = chirps.set_index('date')

# not irrigated areas 

ni = pd.read_csv('not_irrigated_results.csv')
ni = ni.drop(columns=['.geo'])
ni['system:index'] = ni['system:index'].str[:8]
ni['system:index'] = pd.to_datetime(ni['system:index'])
ni = ni.set_index('system:index')

#  irrigated areas 

i = pd.read_csv('irrigated_results.csv')
i = i.drop(columns=['.geo'])
i['system:index'] = i['system:index'].str[:8]
i['system:index'] = pd.to_datetime(i['system:index'])
i = i.set_index('system:index')
```

```python
### merge dataframes - add separate one with precipitation 

all_data = pd.merge(
            i,
            ni,
            how="inner",
            left_index=True,
            right_index=True,
            suffixes=("_i", "_ni"),
            copy=True
        )
all_data_withprecip = pd.merge(
            all_data,
            chirps,
            how="right",
            left_index=True,
            right_index=True
        )

all_data.head()
```

```python
######################## for NDVI

df = all_data_withprecip
df['date'] =  pd.to_datetime(df.index)

# create a figure and axis object
fig, ax1 = plt.subplots(figsize=(15, 6))

# plot the bar chart on the first axis
ax1.bar(df['date'], df['precipitation'], color='royalblue')

# set the y-axis label for the first axis
ax1.set_ylabel('precipitation [mm]', fontsize=12)


# create a second axis object that shares the same x-axis as the first axis
ax2 = ax1.twinx()

# plot the line chart on the second axis

x = date2num(df['date'])
ax2.bar(x, df['NDVI_i'], width=1, color='mediumseagreen')
ax2.bar(x+1, df['NDVI_ni'], width=1, color='goldenrod')

# set the y-axis label for the second axis
ax2.set_ylabel('NDVI', fontsize=12)

# set the x-axis label and title
ax1.set_xlabel('Date', fontsize=12)
ax1.set_title('Development of NDVI June - August 2023', fontsize=14)

ax1.set_ylim([0, 50])
ax2.set_ylim([0, 0.6])

ax1.invert_yaxis()
ax1.xaxis.set_major_locator(ticker.MultipleLocator(base=3))
fig.autofmt_xdate()
# show the plot
plt.show()

plt.savefig('NDVI.png')

```
